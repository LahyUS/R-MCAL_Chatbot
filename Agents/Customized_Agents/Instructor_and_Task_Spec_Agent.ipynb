```python
# pip install transformers
from fnmatch import fnmatch
from pathlib import Path
from typing import Dict, List, Optional, Union
from packaging import version
import os

from huggingface_hub import HfApi, HfFolder, cached_download, hf_hub_url
import huggingface_hub

HUGGINGFACE_HUB_CACHE = "~/.cache/huggingface/transformers"
DEFAULT_CACHE = Path.cwd() / "downloaded-models"

def snapshot_download(
    repo_id: str,
    revision: Optional[str] = None,
    cache_dir: Union[str, Path, None] = None,
    library_name: Optional[str] = None,
    library_version: Optional[str] = None,
    user_agent: Union[Dict, str, None] = None,
    ignore_files: Optional[List[str]] = None,
    use_auth_token: Union[bool, str, None] = None,
) -> str:
    """
	    credit/source: https://github.com/UKPLab/sentence-transformers/blob/master/sentence_transformers/util.py
    Method derived from huggingface_hub.
    Adds a new parameters 'ignore_files', which allows to ignore certain files / file-patterns
    """
    if cache_dir is None:
        cache_dir = HUGGINGFACE_HUB_CACHE
    if isinstance(cache_dir, Path):
        cache_dir = str(cache_dir)

    _api = HfApi()

    token = None
    if isinstance(use_auth_token, str):
        token = use_auth_token
    elif use_auth_token:
        token = HfFolder.get_token()

    model_info = _api.model_info(repo_id=repo_id, revision=revision, token=token)

    storage_folder = os.path.join(cache_dir, repo_id.replace("/", "_"))

    all_files = model_info.siblings
    # Download modules.json as the last file
    for idx, repofile in enumerate(all_files):
        if repofile.rfilename == "modules.json":
            del all_files[idx]
            all_files.append(repofile)
            break

    for model_file in all_files:
        if ignore_files is not None:
            skip_download = False
            for pattern in ignore_files:
                if fnmatch.fnmatch(model_file.rfilename, pattern):
                    skip_download = True
                    break

            if skip_download:
                continue

        url = hf_hub_url(
            repo_id, filename=model_file.rfilename, revision=model_info.sha
        )
        relative_filepath = os.path.join(*model_file.rfilename.split("/"))

        # Create potential nested dir
        nested_dirname = os.path.dirname(
            os.path.join(storage_folder, relative_filepath)
        )
        os.makedirs(nested_dirname, exist_ok=True)

        cached_download_args = {
            "url": url,
            "cache_dir": storage_folder,
            "force_filename": relative_filepath,
            "library_name": library_name,
            "library_version": library_version,
            "user_agent": user_agent,
            "use_auth_token": use_auth_token,
        }

        if version.parse(huggingface_hub.__version__) >= version.parse("0.8.1"):
            # huggingface_hub v0.8.1 introduces a new cache layout. We sill use a manual layout
            # And need to pass legacy_cache_layout=True to avoid that a warning will be printed
            cached_download_args["legacy_cache_layout"] = True

        path = cached_download(**cached_download_args)

        if os.path.exists(path + ".lock"):
            os.remove(path + ".lock")

    return storage_folder
```

```python
######################
#  FILE HANDLER      #
######################
import glob
import pickle
```

```python
is_handle_file = 0
###############################################################################################
#  PROCESS: Extract text from multiple PDF documents and save as .pickle file
###############################################################################################

def extract_pdf(pdf_path, vector_store_path):
  #%cd '/content/drive/MyDrive/Colab_env/Renesas/db/pdf/vector_db/'
  # Read PDF file
  if is_handle_file == 1:
    directory_path = pdf_path
    file_paths = glob.glob(directory_path + "/*")

    number = 0
    for file_path in file_paths:
      print('file_path: ', file_path)
      loader = PyPDFLoader(file_path)
      docs = loader.load()
      texts = text_splitter.split_documents(docs)

      # save the docs variable to a file
      db_path = vector_store_path+ 'docs_' + str(number)+'.pickle'
      print('db_path no.', number, ': ', db_path)
      with open(db_path, 'wb') as f:
          pickle.dump(texts, f)
      number += 1


###############################################################################################
#  PROCESS: Load documents from .pickle file
###############################################################################################
def load_pickle_vector_store(vector_store_path):
  # set the directory path
  directory_path = vector_store_path
  # use glob to get all the file paths in the directory
  file_paths = glob.glob(directory_path + "/*")

  texts = []
  number = 0
  for file_path in file_paths:
    print(file_path)
    with open(file_path, 'rb') as f:
      doc = pickle.load(f)
    texts.extend(doc)
    print(len(texts))
    number += 1
  print('total: ', len(texts), '- type: ', type(texts))
  print('num of doc: ', number)

  return texts
```

```python
from typing import Dict
import re


class Instructor_Agent():
  intructor_chain: LLMChain
  linear_mark_chain: LLMChain
  router_chain: LLMRouterChain
  destination_chains= Dict[str, LLMChain]

  def __init__(self, intructor_chain, linear_mark_chain, router_chain, destination_chains, tools_abstract):
    self.intructor_chain    = intructor_chain
    self.linear_mark_chain  = linear_mark_chain
    self.router_chain       = router_chain
    self.destination_chains = destination_chains
    self.tools_abstract     = tools_abstract


  def parse_instruction(self, instruction):
    # Split the string based on newline character and remove leading/trailing spaces
    steps = [item.strip() for item in instruction.split('\n') if item.strip()]
    return steps

  def parse_mark_list(self, steps):
    # Removing the brackets and splitting the string by comma
    elements = steps[1:-1].split(',')
    # Parsing each element as boolean and creating a boolean list
    marks = [bool(element.strip()) for element in elements]
    return marks

  def find_tool(self, name=str, prompt_infos=[]):
    for p in prompt_infos:
      if p["name"] == name:
        return p["tool_name"]

  def execute_action(self, action_name, action_input, results):
    tool_name = self.find_tool(action_name, prompt_infos)

    if tool_name != "None":
      print(f"***[DEBUG] tool_name: {tool_name}")
      # TODO
      if tool_name == "Linux-Shell":
        # Generate the correct command
        command = self.destination_chains[action_name](action_input)
        print(f"[DEBUG] Command: {command['text']}")
        # Define main_parameter
        previous_content = results[-1]['text']
        parameter = "/content/drive/MyDrive/Colab_env/plan.txt"
        #print(f"[DEBUG] Type: {type(previous_content)} - The previous_content: {previous_content}")
        if "echo" in command['text']:
          # Run command
          correct_command = 'echo "{}" > "{}"'.format(previous_content, parameter)
          print(f"***[DEBUG] The correct command: {correct_command}")
          result = self.tools_abstract[tool_name].func(correct_command)
          results.append(result)

      elif "Chain" in tool_name:
        print(f"***[DEBUG] 1")
        new_action_input = action_input['input']
        print(f"***[DEBUG] Running Chain Tool with new_action_input: {new_action_input}")
        print(f"***[DEBUG] Type(new_action_input): {type(new_action_input)}")

        result = self.tools_abstract[tool_name].func(new_action_input)
        results.append(result)

      else:
        print(f"***[DEBUG] 2")
        result = self.tools_abstract[tool_name].func(action_input)
        results.append(result)

    else:
      print(f"***[DEBUG] 3")
      result = self.destination_chains[action_name](action_input)
      results.append(result)


  def run(self, inputs):
    print("\n\n[INSTRUCTOR] Start Process")
    #steps = self.intructor_chain(inputs)
    #parsed_steps = self.parse_instruction(steps['text'])
    parsed_steps = [item.strip() for item in inputs['instruction'].split(',') if item.strip()]
    linear_marks = self.linear_mark_chain(parsed_steps)
    marks_list = self.parse_mark_list(linear_marks['text'])
    print(f"\n[DEBUG] marks_list: {marks_list}")
    results = []

    for i in range(len(parsed_steps)):
      print(f"[DEBUG] parsed_steps[{i}]: {parsed_steps[i]}")

    print(f"\n[INSTRUCTOR] Do the 1st Step: {parsed_steps[0]}")
    route = self.router_chain.route(parsed_steps[0])
    #print(f"[DEBUG] Route OK")
    action_name = route.destination
    print(f"***[DEBUG] action_name: {action_name}")
    action_input = route.next_inputs
    print(f"***[DEBUG] action_input: {action_input} ")
    self.execute_action(action_name, action_input, results)
    print(f"***[DEBUG] results[-1]: {results[-1]} ")

    for i in range (1, len(parsed_steps)):
      print(f"\n[INSTRUCTOR] Do the {i+1}th step: {parsed_steps[i]}")
      route = self.router_chain.route(parsed_steps[i])
      action_name = route.destination
      print(f"***[DEBUG] action_name: {action_name}")
      if action_name == "Linux-Shell":
          action_input = parsed_steps[i]
      else:
        if marks_list[i] == True:
          action_input = results[-1]
        else: action_input = parsed_steps[i]

      print(f"***[DEBUG] action_input: {action_input}")
      self.execute_action(action_name, action_input, results)

    print("\n\n[INSTRUCTOR] End Process")
    return results
```

```python
class Utils():
  @classmethod
  def remove_substring_from_second_occurrence(cls, string, keyword):
      index = string.find(keyword)  # Find the index of the first occurrence of the keyword
      if index != -1:
          index = string.find(keyword, index + len(keyword))  # Find the index of the second occurrence
          if index != -1:
              substring_to_remove = string[index:]  # Get the substring starting from the second occurrence
              result = string.replace(substring_to_remove, '')  # Remove the substring from the string
              return result
      return string  # Keyword not found or less than two occurrences, return the original string

  @classmethod
  def parse_input(cls, key_agrument_pairs, func_name=str, input_string=str):
    # Convert the input string to lowercase for case-insensitive matching
    input_string = input_string.lower()
    key_arguments = key_agrument_pairs[func_name]
    ret_params = {}

    # Iterate through the keywords list
    for para_num, agruments in enumerate(key_arguments, start=0):
      param_key = 'param' + str(para_num)
      is_has_param = False
      # Iterate through each keyword list
      for keyword in agruments:
        if keyword in input_string:
          is_has_param = True
          print(f"keyword: {keyword}")
          # Extract the issue name
          start_index = input_string.index(keyword) + len(keyword) + 1
          if start_index >= len(input_string):
            ret_params[param_key] = "None"
          else:
            param_val = input_string[start_index:].split()[0]
            ret_params[param_key] = param_val
          #print(f"param_key: {param_key} - param_val: {param_val}")
      if is_has_param == False:
        ret_params[param_key] = "None"

    return ret_params
```

```python
from typing import Dict
import re
import json
from pathlib import Path
from pprint import pprint

class Task_Specific_Agent():
  router_chain: LLMRouterChain
  destination_chains= Dict[str, LLMChain]
  task_spec = {}

  @classmethod
  def handle_task_spec(cls, task_spec_path):
    #file_path= '/content/task_spec.json'
    file = open(task_spec_path, 'r')
    file_contents = file.read()
    data = json.loads(file_contents)
    processed_data = {}
    for task in data['tasks']:
        task_name = task['task_name']
        task_type = task['task_type']
        steps = task['instruction']
        processed_data[task_name] = {'task_type': str, 'steps': []}
        processed_data[task_name]['task_type'] =  task_type

        for step in steps:
            step_instruct = step['step_instruct']
            tool_name = step['tool_name']
            processed_data[task_name]['steps'].append({
                'step_instruct': step_instruct,
                'tool_name': tool_name
            })
    file.close()
    return  processed_data

  def __init__(self, task_spec_qa_run, task_spec_path, router_chain, destination_chains, tools_abstract, agent_router_chain):
    self.task_spec_qa_run   = task_spec_qa_run
    self.task_spec_path     = task_spec_path
    self.router_chain       = router_chain
    self.destination_chains = destination_chains
    self.tools_abstract     = tools_abstract
    self.task_spec          = Task_Specific_Agent.handle_task_spec(self.task_spec_path)
    self.agent_router_chain = agent_router_chain

  @staticmethod
  def parse_step_instruct(input, pattern, replacewith):
    return input.replace(pattern, replacewith)

  def handle_step_instruct(self, step_instruct, request, results):
    is_replace_with_prev_result = False
    if "{replacement}" in step_instruct:
      parsed_step_instruct = self.parse_step_instruct(step_instruct, '{replacement}', request)
      return is_replace_with_prev_result, parsed_step_instruct

    elif "{prev_result_as_input}" in step_instruct:
      is_replace_with_prev_result = True
      parsed_step_instruct = self.parse_step_instruct(step_instruct, '{prev_result_as_input}', "")
      return is_replace_with_prev_result, parsed_step_instruct

    return is_replace_with_prev_result, step_instruct

  def find_tool(self, action_name=str):
    for tool_name, tool_func in self.tools_abstract.items():
      if tool_name == action_name:
        return True
    return False

  @staticmethod
  def parse_task_name(sentence):
    pattern = r'"(.*?)"'  # Regular expression pattern to match the keywords enclosed in double quotes
    matches = re.findall(pattern, sentence)
    if matches:
        return matches[0]  # Return the first match
    else:
        return None  # Return None if no match found

  def execute_action(self, action_name, action_input, results):
    is_execute_tool = self.find_tool(action_name)
    if is_execute_tool == True:
      print(f"***[DEBUG] tool_name: {action_name}")
      # TODO
      if action_name == "Linux-Shell":
        # Generate the correct command
        command = self.destination_chains[action_name](action_input)
        print(f"[DEBUG] Command: {command['text']}")
        # Define main_parameter
        previous_content = results[-1]['text']
        parameter = "/content/drive/MyDrive/Colab_env/plan.txt"
        #print(f"[DEBUG] Type: {type(previous_content)} - The previous_content: {previous_content}")
        if "echo" in command['text']:
          # Run command
          correct_command = 'echo "{}" > "{}"'.format(previous_content, parameter)
          print(f"***[DEBUG] The correct command: {correct_command}")
          result = self.tools_abstract[action_name].func(correct_command)
          results.append(result)

      elif "Chain" in action_name:
        new_action_input = action_input['input']
        result = self.tools_abstract[action_name].func(new_action_input)
        results.append(result)

      else: # if action_name != Linux-Shell && != Chain
        result = self.tools_abstract[action_name].func(action_input)
        results.append(result)

    else: # If use LLM model answer based on Prompt Template
      result = self.destination_chains[action_name](action_input)
      results.append(result)

  def run(self, inputs):
    print("\n\n[TASK_SPEC_AGENT] Start Process")
    request = inputs['instruction']
    #prefix = "In the pre-defined tasks in MCAL development, determin which task is suitable for this request:\n <<REQUEST>>: "
    #posfix = "\nYou just have to answer the name of the suitable task. Do not generate a new task name"
    #find_task = prefix + request + posfix
    #suit_task = self.task_spec_qa_run(find_task)
    #print(f"***[DEBUG] suit_task: {suit_task}")
    #task_name = self.parse_task_name(suit_task)
    task_route = self.agent_router_chain.route(request)
    task_name = task_route.destination
    print(f"[TASK_SPEC_AGENT] Suitable Task: {task_name}")

    if task_name not in self.task_spec:
      print("[TASK_SPEC_AGENT] I cannot find the suitable task for your request.\nPlease give more information about your requirement context!")
      return ""

    task_type = self.task_spec[task_name]['task_type']
    print(f"***[DEBUG] task_type: {task_type}")

    results = []
    if task_type == 'complex':
      if(task_name == 'Jira Info Collector'):
        jira = Jira_Info_Collector()
        username = input('Enter your username: ')
        password = input('Enter your password: ')
        access_token = "NjQ0Mzg1MDcyMTAxOtTUPo2ejc820FtHX64OdMz1x/dj"
        jira.jira_auth(username=username, password=password, token=access_token)
        #result = jira.run(request)
        result = jira.run(request)
        results.append(result)

      elif(task_name == 'Confluence Info Collector'):
        confluence = Confluence_Info_Collector()
        username = input('Enter your username: ')
        password = input('Enter your password: ')
        #access_token = "NjQ0Mzg1MDcyMTAxOtTUPo2ejc820FtHX64OdMz1x/dj"
        confluence.confluence_auth(username=username, password=password)#, token=access_token)
        #result = jira.run(request)
        result = confluence.run(request)
        results.append(result)

      elif(task_name == 'Run MCAL'):
        MCAL_Operation()

    else:
      # AGENT DO THE REMAIN STEPS
      task_info = self.task_spec[task_name]
      steps = task_info['steps']
      for step_num, step in enumerate(steps[:], start=1):
        raw_step_instruct = step['step_instruct']
        is_replace_with_prev_result, step_instruct = self.handle_step_instruct(raw_step_instruct, request, results)

        route = router_chain.route(step_instruct)
        if is_replace_with_prev_result == True:
          action_input = results[-1]
        else: action_input = route.next_inputs

        print(f"\n***[DEBUG] Do the {step_num}th step: {step_instruct}")
        action_name = route.destination
        print(f"***[DEBUG] action_name: {action_name}")
        print(f"***[DEBUG] action_input: {action_input} ")

        self.execute_action(action_name, action_input, results)

    print("\n\n[TASK_SPEC_AGENT] End Process")
    return results
```

```python
# Set up a prompt template
from langchain.prompts import BaseChatPromptTemplate
from langchain.schema import AgentAction, AgentFinish, HumanMessage

class CustomPromptTemplate(BaseChatPromptTemplate):
    # The template to use
    template: str
    # The list of tools available
    tools: List[Tool]

    def format_messages(self, **kwargs) -> str:
        # Get the intermediate steps (AgentAction, Observation tuples)
        # Format them in a particular way
        intermediate_steps = kwargs.pop("intermediate_steps")
        thoughts = ""
        for action, observation in intermediate_steps:
            thoughts += action.log
            thoughts += f"\nObservation: {observation}\nThought: "
        # Set the agent_scratchpad variable to that value
        kwargs["agent_scratchpad"] = thoughts
        # Create a tools variable from the list of tools provided
        kwargs["tools"] = "\n".join([f"{tool.name}: {tool.description}" for tool in self.tools])
        # Create a list of tool names for the tools provided
        kwargs["tool_names"] = ", ".join([tool.name for tool in self.tools])
        formatted = self.template.format(**kwargs)
        return [HumanMessage(content=formatted)]
```

```python
prompt = CustomPromptTemplate(
    template=template,
    tools=tools,
    # This omits the `agent_scratchpad`, `tools`, and `tool_names` variables because those are generated dynamically
    # This includes the `intermediate_steps` variable because that is needed
    input_variables=["input", "intermediate_steps"]
)
```

```python
class CustomOutputParser(AgentOutputParser):

    def parse(self, llm_output: str) -> Union[AgentAction, AgentFinish]:
        # Check if agent should finish
        if "Final Answer:" in llm_output:
            return AgentFinish(
                # Return values is generally always a dictionary with a single `output` key
                # It is not recommended to try anything else at the moment :)
                return_values={"output": llm_output.split("Final Answer:")[-1].strip()},
                log=llm_output,
            )
        # Parse out the action and action input
        regex = r"Action\s*\d*\s*:(.*?)\nAction\s*\d*\s*Input\s*\d*\s*:[\s]*(.*)"
        match = re.search(regex, llm_output, re.DOTALL)
        if not match:
            raise ValueError(f"Could not parse LLM output: `{llm_output}`")
        action = match.group(1).strip()
        action_input = match.group(2)
        # Return the action and action input
        return AgentAction(tool=action, tool_input=action_input.strip(" ").strip('"'), log=llm_output)

output_parser = CustomOutputParser()
```
